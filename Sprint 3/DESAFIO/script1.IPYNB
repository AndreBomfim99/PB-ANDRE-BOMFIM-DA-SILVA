CÓDIGO 32: Este código cria uma função para formatar os valores no eixo y em milhões como unidade adicionando o M para indicar milhões, selecionar os aplicativos com 1 bilhão de instalações, criar o gráfico de barras com ajustes na escala do eixo y, aplica o desempate baseado na coluna Reviews, aplica a formatação no eixo y, ajusta o limite do eixo y com uma margem de 10%.

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def milhoes(x, pos):
    return f'{x * 1e-6:.0f}M'  

top_5_instalados = dados_google_play[dados_google_play['Installs'] == 1_000_000_000]
top_5_desempatados = top_5_instalados.nlargest(5, 'Reviews')[['App', 'Installs', 'Reviews']]

plt.figure(figsize=(10, 6))
plt.bar(top_5_desempatados['App'], top_5_desempatados['Reviews'])
plt.title('Top 5 Aplicativos com 1 Bilhão de Instalações (Desempatados por Número de Avaliações)')
plt.xlabel('Aplicativos')
plt.ylabel('Número de Reviews (em Milhões)')

plt.gca().yaxis.set_major_formatter(FuncFormatter(milhoes))
plt.ylim(0, top_5_desempatados['Reviews'].max() * 1.1)  
plt.xticks(rotation=45)
plt.show()
```

CÓDIGO 33: Contagem e separação das categorias, onde calcula a frequência de cada categoria e define as 5 principais, somando as restantes como "Outros". Combina as categorias com pd.concat para unir as 5 principais categorias com a nova categoria "Outros". Refina o gráfico de pizza onde destaca as categorias principais, usando bordas nas fatias e um efeito de "explosão" para melhorar a visualização.

```python
import matplotlib.pyplot as plt
import pandas as pd

frequencia_categorias = dados_google_play['Category'].value_counts()
top_categorias = 5
categorias_principais = frequencia_categorias[:top_categorias]
categorias_outras = frequencia_categorias[top_categorias:].sum()
categorias_final = pd.concat([categorias_principais, pd.Series([categorias_outras], index=['Outros'])])
plt.figure(figsize=(10, 8))
plt.pie(categorias_final, labels=categorias_final.index, autopct='%1.1f%%', startangle=140, 
        wedgeprops={'edgecolor': 'black'}, pctdistance=0.85, explode=[0.1]*len(categorias_final))
plt.title('Distribuição de Categorias de Aplicativos (Principais e Outros)')
plt.show()
```

CÓDIGO 42: Converte as colunas Price e Installs para tipos numéricos, substituindo valores inválidos por NaN para facilitar o agrupamento. Cria as faixas de preço definindo intervalos de preço (bins) e rótulos (labels) e cria a coluna Faixa_Preco que classifica cada aplicativo em uma faixa específica. 
Calcula a média de Reviews e Installs por faixa de preço, gerando media_reviews_por_preco e media_installs_por_preco.
Combina o gráfico de barras usando ax1.bar para exibir a média de instalações por faixa de preço, e usa as linhas com ax2.plot para exibir a média de avaliações na mesma faixa de preço, compartilhando o eixo x. O restante das linhas de código são apenas a configuração de títulos e rótulos do g´rafico.  

```python
import pandas as pd
import matplotlib.pyplot as plt
dados_google_play['Price'] = pd.to_numeric(dados_google_play['Price'], errors='coerce')
dados_google_play['Installs'] = pd.to_numeric(dados_google_play['Installs'], errors='coerce')
bins = [0, 5, 10, 20, 50, float('inf')]
labels = ['0-5', '5-10', '10-20', '20-50', '50+']
dados_google_play['Faixa_Preco'] = pd.cut(dados_google_play['Price'], bins=bins, labels=labels)

media_reviews_por_preco = dados_google_play.groupby('Faixa_Preco', observed=True)['Reviews'].mean()
media_installs_por_preco = dados_google_play.groupby('Faixa_Preco', observed=True)['Installs'].mean()
print("Média de Avaliações por Faixa de Preço:")
print(media_reviews_por_preco)
print("\nMédia de Instalações por Faixa de Preço:")
print(media_installs_por_preco)

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(media_installs_por_preco.index, media_installs_por_preco.values, color='lightblue', label='Média de Instalações')
ax1.set_ylabel('Média de Instalações (Milhões)')
ax1.tick_params(axis='y')
ax1.set_xlabel('Faixa de Preço ($)')
ax1.set_ylim(0, media_installs_por_preco.max() * 1.2) 
ax2 = ax1.twinx()
ax2.plot(media_reviews_por_preco.index, media_reviews_por_preco.values, color='darkblue', marker='o', linestyle='-', label='Média de Avaliações')
ax2.set_ylabel('Média de Avaliações')
ax2.tick_params(axis='y', labelcolor='darkblue')
ax2.set_ylim(0, media_reviews_por_preco.max() * 1.2)
plt.title('Média de Avaliações e Instalações por Faixa de Preço')
fig.tight_layout()
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9), bbox_transform=ax1.transAxes)
plt.show()
```

CÓDIGO 43: Este código usa dados_faltantes = dados_google_play.isnull().sum() para calcular a quantidade de valores ausentes por coluna, com
dados_faltantes = dados_faltantes[dados_faltantes > 0] filtra apenas colunas com valores ausentes, o restante do código desenha o gráfico de barras.

```python
import pandas as pd
import matplotlib.pyplot as plt

dados_google_play = pd.read_csv('googleplaystore.csv')
dados_google_play = dados_google_play.drop(columns=['Faixa_Preco'], errors='ignore')
dados_faltantes = dados_google_play.isnull().sum()
dados_faltantes = dados_faltantes[dados_faltantes > 0]
plt.figure(figsize=(10, 6))
dados_faltantes.plot(kind='bar', color='salmon')
plt.title('Quantidade de Valores Ausentes por Coluna')
plt.xlabel('Colunas')
plt.ylabel('Quantidade de Valores Ausentes')
plt.show()
```
